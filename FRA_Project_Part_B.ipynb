## Visualizing Returns and Volatility

```python
data = pd.DataFrame({
    'Average Return': Return_of_Stocks.mean(),
    'Volatility': Return_of_Stocks.std()
})  ## Complete the code to get a dataframe for mean  and std. deviation for the returns of all stocks
data
```

```python
import seaborn as sns

fig = plt.figure(figsize=(9,5))
ax = fig.add_subplot(111)
sns.scatterplot(x='Volatility', y='Average Return', data=data, s=100)

plt.scatter(data['Volatility'], data['Average Return'], s=100)  ## Complete the code to get the plot of returns vs volatility for the returns for all stocks
plt.axhline(y=0,linestyle='--', color = "red")

for index, row in data.iterrows():
    ax.text(row['Volatility'] + 0.001, row['Average Return'] + 0.0002, index)
plt.xlabel('Volatility')
plt.ylabel('Net returns')
plt.title('Net return vs Volatility for stocks')
plt.grid()
plt.show()
```
